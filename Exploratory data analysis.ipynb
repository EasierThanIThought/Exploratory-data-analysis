## V. Evaluation


### 1. Linear Regression (Regresja Liniowa)
#### Charakterystyka:

Regresja liniowa jest podstawową techniką modelowania predykcyjnego, która zakłada liniową zależność między cechami a wartością docelową.
Model stara się dopasować prostą linię (hiperplan w przypadku wielu cech) do danych, minimalizując sumę kwadratów błędów (różnicy między wartościami rzeczywistymi a przewidywanymi).

#### Zasadność zastosowania:

Regresja liniowa jest odpowiednia, gdy istnieje liniowa zależność między zmiennymi wejściowymi (cechami) a zmienną docelową.
Jest prostym i szybkim modelem, dobrze sprawdzającym się na małych i średnich zestawach danych.


### 2. Decision Tree (Drzewo Decyzyjne)
#### Charakterystyka:

Drzewo decyzyjne jest modelem, który iteracyjnie dzieli dane na mniejsze podzbiory w oparciu o wartości cech, tworząc strukturę drzewiastą. Każdy węzeł w drzewie reprezentuje decyzję opartą na wartościach cech, a każdy liść reprezentuje wartość docelową.
Model jest łatwy do zrozumienia i wizualizacji, co ułatwia interpretację wyników.


#### Zasadność zastosowania:

Drzewa decyzyjne są skuteczne w modelowaniu złożonych, nieliniowych zależności między cechami a wartością docelową.
Dobrze radzą sobie z danymi kategoricznymi i mogą modelować interakcje między cechami bez konieczności wcześniejszego przygotowania danych.

### 3. Random Forest (Las Losowy)
#### Charakterystyka:

Random Forest to zespół (ensemble) wielu drzew decyzyjnych, trenowanych na losowych podzbiorach danych i cech. Wynik końcowy jest średnią predykcji wszystkich drzew.
Model wykorzystuje podejście bootstrap aggregating (bagging) i losowy wybór cech, co poprawia dokładność i zmniejsza ryzyko nadmiernego dopasowania.
#### Zasadność zastosowania:

Random Forest jest skuteczny dla dużych zestawów danych i skomplikowanych zależności między cechami a wartością docelową.
Jest bardziej odporny na nadmierne dopasowanie niż pojedyncze drzewo decyzyjne i zwykle osiąga lepsze wyniki.


Wyniki ewaluacji można przedstawić w tabelach wynikowych, które pozwalają na łatwe porównanie różnych modeli w kontekście zarówno upload jak i download. Tabele te powinny zawierać wartości MSE (Mean Squared Error), MAE (Mean Absolute Error) oraz R2 (R-squared) dla każdego modelu, co pozwoli na ocenę ich wydajności.
